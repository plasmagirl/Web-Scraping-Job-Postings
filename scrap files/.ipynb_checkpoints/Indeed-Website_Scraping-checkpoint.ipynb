```python
import requests
from bs4 import BeautifulSoup
```

```python
#I am going to parse the Indeed website using Data Scientist and Remote as the search parameters
url = 'https://www.indeed.com/q-data-scientist-l-remote-jobs.html'

#request library downloads the contents of the website that can be accessed by the content object
response = requests.get(url)
#response.content
```

```python
#BeautifulSoup library parses the contents of the website. This library allows us to extract tags from an HTML document.
parser = BeautifulSoup(response.content, 'html.parser')
#print(parser.prettify())
```

```python
parser.find_all('tbody')[0].get_text()
```

```python
#to access the contents of body tag
body = parser.body
body
```

```python
body_content = body.find(id='resultsBodyContent')
print(body_content.prettify())
```

```python
job_search = body_content.find_all('div', class_ ="jobsearch-SerpJobCard unifiedRow row result")
for i in range(0,14):
    print(job_search[i].find(class_='company').text.strip())
    print(job_search[i].find(class_='recJobLoc')['data-rc-loc'])
    print(job_search[i].find(class_='remote'))
    try:
        print(job_search[i].find(class_='salary').text.strip())
    except AttributeError:
        print('Salary not available')    
    print(job_search[i].find(class_='summary').text.strip())
    print('******')
```

```python
for js in job_search:
    title = js.find('h2', class_='title')
   # company = js.find('div', class_='sjcl') #this tag & class gives company, location and company rating info
    company = js.find(class_='company') #only company info
    location = js.find('div', class_='recJobLoc' ) #only location, but stored under 'data-rc-loc', not text
    location_remote = js.find('span', class_='remote')
    salary = js.find('span', class_='salaryText')
    snippet = js.find(class_='summary')
    if None in (title, company, location, location_remote): #if any of the information is missing, don't print 
        continue
    print(title.text.strip()) #only text, strip() strips all empty area around the text
    print(company.text.strip())
    print(location['data-rc-loc'].strip())
    print(location_remote.text.strip())
    try:
        print(salary.text.strip())
    except AttributeError:
        print('Salary not available')
    print(snippet.text.strip())
    print('  ')
```
